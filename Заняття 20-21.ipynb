```python
import numpy as np


def sigmoid(x):
    return 1/(1+np.exp(-x))

class Neuron:
    
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
        
    def feedforward(self, inputs):
        # Weights * inputs + bias, applying activation function
        total = np.dot(self.weights, inputs) + self.bias 
        return sigmoid(total)

weights = np.array([0,1])
bias = 4

n = Neuron(weights, bias)

x = np.array([2,3])

print(n.feedforward(x))
```

```python
class LayerDefinition:
    
    def __init__(self):
        
        weights = np.array([0,1])
        bias = 4
        
        # Neuron classes
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias)
        
        self.k1 = Neuron(weights, bias)
        
    def feedforward(self, x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        
        # The inputs for out_k1 are outputs from h1 and h2
        out_k1 = self.k1.feedforward(np.array([out_h1, out_h2]))
        
        return out_k1

network = LayerDefinition()
x = np.array([2,3])
print(network.feedforward(x))
```

```python
import tensorflow as tf
# Tensorflow 2
from tensorflow import keras
```

```python
model = keras.Sequential()

input_layer = keras.layers.Dense(4, input_shape=[4], activation='sigmoid')
model.add(input_layer)
# activation functions -- Лайфхак - сігмоїд використовують для аутпутів від 0 до 1,
# від -1 до 1 -- гіперболічний тангенс tanh 
# від 0 до +infty використовують relu (leaky relu)

output_layer = keras.layers.Dense(1, activation='softmax')
model.add(output_layer)
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
```

```python
import pandas as pd

df = pd.read_excel('data.xls')
```

```python
df = df[df['species'] != 'Iris-whatever']
df[df.iloc[:, :4] <= 0] = np.nan
df = df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(df.species.unique())
```

```python
encoded_species = le.fit_transform(df.species)
df['encoded_species'] = encoded_species
df
```

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:,:4].to_numpy()
Y = df.encoded_species.to_numpy()

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
history = model.fit(X_train, Y_train,
                   epochs=500,
                   steps_per_epoch = 10,
                   validation_split = 0.25,
                   batch_size = 40)
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print('Test score: ', loss)
print('Test accuracy: ', accuracy)
```
