```python
import sqlite3
import pandas as pd
```

```python
connection=sqlite3.connect('database.db')
connection.execute('PRAGMA foreign_keys = ON')
cursor = connection.cursor()
```

```python
sql_query = '''
CREATE TABLE Table1 (
    id TEXT PRIMARY KEY,
    vehicle_type TEXT,
    mileage INTEGER,
    max_speed INTEGER,
    passenger_count INTEGER,
    category TEXT
    )'''
```

```python
cursor = connection.cursor()
try:
    cursor.execute(sql_query)
    connection.commit()
except Exception as e:
    print(e)
    connection.rollback()
```

```python
import random

id_ = [i for i in range(20)]
vehicle_type = ['car' for i in range(12)]+['truck' for j in range(5)]+['motorcycle' for k in range(3)]
mileage = [random.randrange(10000, 350000) for i in range(20)]
max_speed = [random.randrange(170000, 220000) for i in range(20)]
passenger_count = ['4' for i in range(7)]+['6' for j in range(3)]+['2' for k in range(2)]+['3' for l in range(3)]+['6' for m in range(2)]+['2' for n in range(2)]+[1 for o in range(1)]
category = ['B' for i in range(12)]+['C' for j in range(5)]+['A' for k in range(3)]
dataframe = pd.DataFrame(list(zip(id_,vehicle_type,mileage,max_speed,passenger_count, category)),
                        columns=['id', 'vehicle_type', 'mileage', 'max_speed', 'passenger_count', 'category'])
dataframe
```

```python
dataframe.to_sql('Table1', connection, if_exists = 'replace', index = False)
connection.commit()
```

```python
select='''
SELECT *
FROM Table1
'''
```

```python
pd.read_sql_query(select, connection)
```

```python
sql_query2 = '''
CREATE TABLE Table2 (
    id TEXT PRIMARY KEY,
    vehicle_id TEXT
    driver_name TEXT,
    driver_category TEXT,
    FOREIGN KEY (vehicle_id) REFERENCES Table1(id)  
    )'''
```

```python
cursor = connection.cursor()
try:
    cursor.execute(sql_query2)
    connection.commit()
except Exception as e:
    print(e)
    connection.rollback()
```

```python
dataframe.to_sql('Table2', connection, if_exists = 'replace', index = False)
connection.commit()
```

```python
select='''
SELECT *
FROM Table2
'''
```

```python
sql_query3 = '''
CREATE TABLE Table3 (
    id TEXT PRIMARY KEY,
    category TEXT
    permitted_roads TEXT
    )'''
```

```python
cursor = connection.cursor()
try:
    cursor.execute(sql_query3)
    connection.commit()
except Exception as e:
    print(e)
    connection.rollback()
```

```python
id_ = [uuid.uuid4().hex for i in range(10)]
category = ['B' for i in range(4)]+['C' for j in range(3)]+['A' for k in range(3)]
permitted_roads = ['A1' for i in range(3)]+['M1' for j in range(3)]+['M2' for k in range(2)]+['A22' for l in range(1)]+['C16' for m in range(1)]
random.shuffle(permitted_roads)
dataframe = pd.DataFrame(list(zip(id_,category, permitted_roads)),
                        columns=['id', 'category', 'permitted_roads'])
dataframe
```

```python
dataframe.to_sql('Table3', connection, if_exists = 'replace', index = False)
connection.commit()
```

```python
select='''
SELECT *
FROM Table3
'''
```

```python
pd.read_sql_query(select, connection)
```

```python
sql_query = pd.read_sql_query('''
SELECT *
FROM Table1 A
JOIN Table2 B ON B.vehicle_id=A.id
WHERE A.category=B.driver_category
''', connection)
df = pd.DataFrame(sql_query, columns = ['id', 'vehicle_type', 'mileage','max_speed','passenger_count', 'category', 'id', 'vehicle_id', 'driver_name','driver_category'])
df
```

```python
select='''
SELECT B.id, B.vehicle_id, B.driver_name, B.driver_category, COUNT (C.permitted_roads) as 'road_count'
FROM Table1 A
JOIN Table2 B ON B.vehicle_id=A.id
JOIN Table3 C ON C.category=A.category
WHERE A.category=B.driver_category 
GROUP BY B.driver_name
HAVING COUNT (C.permitted_roads)>2 
'''

pd.read_sql_query(select, connection)
```

```python
sql_query = pd.read_sql_query('''
SELECT B.id, B.vehicle_id, B.driver_name, B.driver_category, COUNT (C.permitted_roads) as 'road_count'
FROM Table1 A
JOIN Table2 B ON B.vehicle_id=A.id
JOIN Table3 C ON C.category=A.category
WHERE A.category=B.driver_category 
GROUP BY B.driver_name
HAVING COUNT (C.permitted_roads)>2
''', connection)

df = pd.DataFrame(sql_query, columns = ['id', 'vehicle_id', 'driver_name','driver_category', 'road_count'])
df
```

```python
df.to_csv('set2.csv')
```
